```python
all_cities = ["Bangkok", "Paris", "Seattle", "Chicago", "Sapporo", "Taipei", "Mexico", "Tokyo", 
"Denver", "Antwerp", "Barcelona", "Amsterdam", "Zagreb", "Toronto", "Adelaide", "Madrid",
"Athens", "Berlin", "Milan", "Hanoi", "Glasgow", "Edinburgh", "Bogota", "SanJose", "Budapest", 
"Yokohama", "Zurich", "Bern", "Washington", "Boston", "Miami", "Atlanta", "Austin", "Houston",
"Buenos Aires", "Santiago", "Auckland", "Brisbane", "Sydney", "Melbourne", "KualaLumpur", "Taichung", 
"Kowloon", "Chiang Mai", "Johannesburg", "Campinas", "Belo Horizonte", "Sao Paulo", "Phoenix", "Singapore"]
networks = {}
intersections = {}
streets = {}
for city in all_cities:
    m = urb.Map(city, zoom=11)
    if not os.path.exists("/home/srg/projects/urbanity/data/"+city.lower()+'.geojson'):
        continue
    m.add_polygon_boundary("/home/srg/projects/urbanity/data/"+city.lower()+'.geojson')
    networks[city], intersections[city], streets[city] = m.get_street_network(location=city, 
                                               graph_attr=True,
                                               building_attr=True,
                                               catchment_attr=True,
                                               pop_attr=True, 
                                               svi_attr=True,
                                               poi_attr=True,
                                               bandwidth=100,
                                               edge_attr=True,
                                               get_precomputed=True)
```
